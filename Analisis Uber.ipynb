```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns
```

```python
df = pd.read_csv('cleanedd_ata_uber.csv')
df.head(50)
```

```python
print(df.columns)
```

```python
df.info()
```

```python
df=df.drop(columns=['Unnamed: 0'])
```

```python
df['date']= pd.to_datetime(df['date'], format='%Y-%m-%d').dt.date
df['time']= pd.to_datetime(df['time'], format='%H:%M:%S').dt.time
df['month']= pd.to_datetime(df['date']).dt.strftime('%B')
df['day']= pd.to_datetime(df['date']).dt.day_name()
df['time_period']= pd.to_datetime(df['time'], format='%H:%M:%S').dt.hour.apply( lambda x: 'Dawn' 
                                                                               if 0 <= x < 5 
                                                                               else ('Morning' 
                                                                               if 5 <= x < 12 
                                                                               else ('Afternoon'
                                                                               if 12 <= x < 17 
                                                                               else ('Evening' 
                                                                               if 17 <= x < 21
                                                                              else 'Night'))))

bulan_order = ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember']
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
time_period_order = ['Dawn', 'Morning', 'Afternoon', 'Evening', 'Night']

df['month'] = pd.Categorical(df['month'], categories=bulan_order, ordered=True)
df['day'] = pd.Categorical(df['day'], categories=day_order, ordered=True)
df['time_period'] = pd.Categorical(df['time_period'], categories=time_period_order, ordered=True)
df.head(50)
```

```python
df.info()
```

```python
TotalBooking = df['booking_id'].count()
booking_sukses = df[df['booking_status'] == 'Completed']['booking_status'].value_counts()
TotalKeuntungan = df['booking_value'].sum()
Jaraktempuh = df['ride_distance'].sum()
print('Jumlah Transaksi:',TotalBooking)
print('Total Pendapatan:', TotalKeuntungan)
print('Transaksi Sukses:', booking_sukses)
print('Jarak Tempuh:', Jaraktempuh)
```

```python
bookingbulan = df.groupby('month')['booking_id'].count().reindex(bulan_order)
avgbookmont = bookingbulan.mean()
print('Rata-rata Booking per Bulan:', avgbookmont)

pendapatanbulan = df.groupby('month')['booking_value'].sum().reindex(bulan_order)
avgpendapatanmont = pendapatanbulan.mean()
print('Rata-rata Pendapatan per Bulan:', avgpendapatanmont)

succesbulan = df[df['booking_status'] == 'Completed'].groupby('month')['booking_status'].count().reindex(bulan_order)
avgsuccesmont = succesbulan.mean()
print('Rata-rata Transaksi Sukses per Bulan:', avgsuccesmont)

jarakbulan = df.groupby('month')['ride_distance'].sum().reindex(bulan_order)
avgjarakmont = jarakbulan.mean()
print('Rata-rata Jarak Tempuh per Bulan:', avgjarakmont)
```

```python
bulanan = df.pivot_table(index='month', values=['booking_id','booking_value'], aggfunc={'booking_id':'count','booking_value':'sum'}).reindex(bulan_order)
avgbulan = bulanan.mean()
print('Rata-rata Booking dan Pendapatan per Bulan:' , avgbulan)    
print(bulanan)



ax = bulanan.plot(kind='line',secondary_y='booking_value', legend=True, marker='o', figsize=(10,6))
plt.title('Jumlah Booking dan Pendapatan per Bulan')
ax.set_xlabel(None)
ax.set_ylabel('Jumlah Booking', fontsize=12)
ax.right_ax.set_ylabel('Total Pendapatan', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
bulan = df.groupby('date')[['booking_id','booking_value','ride_distance']].agg({'booking_id':'count','booking_value':'sum','ride_distance':'sum'})
bulan.index = pd.to_datetime(bulan.index)
daily_avg = bulan.groupby(bulan.index.to_period('M')).mean().round(0)
avg_daily_avg = daily_avg.mean()

ridemean = df['ride_distance'].mean()
print('Jarak Tempuh/booking :', ridemean)
print('Rata-rata Harian:' , avg_daily_avg)
print(daily_avg)

daily_avgbt = daily_avg.sort_values(by='booking_id', ascending=False)
daily_avgbb = daily_avg.sort_values(by='booking_id', ascending=True)
print('Puncak Booking:', daily_avgbt.head(1))
print('Lembah Booking:', daily_avgbb.head(1))



daily_avgvt = daily_avg.sort_values(by='booking_value', ascending=False)
daily_avgvb = daily_avg.sort_values(by='booking_value', ascending=True)
print('Puncak Pendapatan:', daily_avgvt.head(1))

print('Lembah Pendapatan:', daily_avgvb.head(1))
```

```python

daily_avg.plot(kind='line', secondary_y='booking_value', legend=True, marker='o',figsize=(10,6))
plt.ylabel('Rata-rata Pendapatan Harian')
plt.title('Rata-rata Performa Harian per Bulan')

plt.show()
```

Pendapatan tertinggi terjadi pada bulan Maret
Penggunaan tertinggi terjadi pada bulan Juli
Sedangkan pendapatan dan penggunaan terendah terjadi pada bulan Febuari

```python
pivot = df.pivot_table(index='vehicle_type', values='booking_id', columns='booking_status', aggfunc='count', fill_value=0)
pivot_percentage = pivot.div(pivot.sum(axis=1), axis=0)*100
pivot_percentage = pivot_percentage.round(2)
print(pivot_percentage)

sns.heatmap(pivot_percentage, annot=True,fmt='12g',cmap='coolwarm',xticklabels=True)
plt.title('Peresnase Status Boooking berdasarkan Tipe Kendaraan')
plt.xlabel(None)
plt.ylabel(None)
plt.xticks(rotation =30)  
plt.show()
```

Tingkat penyelesaian secara umum relatif sama berada dikisaran 61-62% di seluruh tipe kendaraan
Masalah pembatalan paling besar terjadi karena Pembatalan oleh Driver dengan rentang 17-18% di seluruh tipe kendaraan sedangkan pembatalan oleh Cuustomer berada di rentang 6-7% di seluruh tipe kendaraan
Perjalanan yang tidak terselesaikan berada pada rentang 5-6% dan
Tidak mendapatkan Driver berada pada 6-7%

Meskipun secara keseluruhan tipe kendaraan memiliki performa yang relatif sama, namun terdapat sedikit permasalahan dimana pada tipe kendaraan Go Sedan memiliki tingkat penyelesaian paling rendah serta pembatalan oleh driver dan tidak menemukan driver berada ditingkat paling tinggi.
